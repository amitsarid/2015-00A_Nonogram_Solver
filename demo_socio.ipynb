### Example

Say you scored 6.56/7 based on 16 ranking soldiers, with a standard deviation of 1.46, all with an accuracy of 2 decimals. 

```python
print(socio.socio_results(6.56, 1.46, 16, plot=False)['result'])
print(socio.socio_results(6.93, 0.25, 15, plot=False)['result'])
```

Thus you can tell that there's only one possible ranking vectors. 14 Soldiers ranked you a 7/7, 1 soldier ranked you 6/7, and 1 soldier ranked you 1/7. 
Without the rogue soldier your score whould have been 6.93/7 with an std of 0.25.
